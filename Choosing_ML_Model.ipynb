# Estimator analysis

Image source: https://scikit-learn.org/stable/tutorial/machine_learning_map/index.html

Some consideration while deciding the algorithm:

1: How much data we have? If we have less data we might need to fetch more.

2: Identify the type of problem. Classification or Regression?

3: See whether the data is labelled or unlabeled?

4: What is the size of the training Dataset

# Classification Algorithms

Lets say the problem statement is to identify categories. eg. whether an email is spam or not. We have labelled data, i.e we have a target/dependent variable present in data

Independent variables (input) ----> Classification model -----> categorical output


**Scenario 1: data < 100,000 samples**

1. Linear SVC (Support vector classifier algo)
2. If Linear SVC doesn't give expected results, check if data is text or not.
   - if text: Naive Bayes algorithm for classification purposes since this algorithm is used for classifying sentiments of users to perform Sentiment Analysis
   - else: KNN i.e, K Nearest Neighbor classifier. If somehow the KNN algorithm doesnt work or gives the right results or accuracy, then we should try SVC i.e, Support Vector Classifier or Ensemble Classifier.


**Scenario 2: data > 100,000 samples**
1. SGD i.e, Stochastic Gradient Descent Classifier.
2. If doesnt perform,  kernel approximation classifier.

# Clustering Algorithms

No target or dependent variable available in the dataset then we will fall into the clustering

Raw data --> Clustering algo --> Output

**Scenario 1: Number of categoris is known and data < 10,000**

1. Opt fo K-means algo.
2. If it doesn't perform well use Spectral Clustering or GMM i.e, Gaussian Mixture Model clustering.

**Scenario 2: Number of categoris is known and data > 10,000**

Use mini-batch k-means clustering

**Scenario 3: Number of categoris is not known and data < 10,000**

Mean Shift or VBGMM i.e, Variational Bayesian Gaussian Mixture model.

**Scenario 3: Number of categoris is not known and data > 10,000**

Difficult scenario


# Regression Algorithms

If we are not predicting a category and predicting a quantity then we will fall into regression

**Scenario 1: data > 100,000 samples**

SGD i.e, Stochastic Gradient Descent Regressor

**Scenario 2: data < 100,000 samples and few feature needed**

Lasso or ElasticNet regression

**Scenario 3: data < 100,000 samples and more feature needed**

1. Ridge regression or SVR i.e, Support Vector Regressor with kernel defined as linear. (graph will be a straight line)
2. If it doesnt work, use SVR with RBF i.e, Radial Basis Function (he graph will be curvilinear in nature)
3. Also check ensemble regressor or an ensemble regression algorithm.


# Dimensionality Reduction Algorithms

If we are not predicting a quantity but only looking at the features/ independent variables, you wanted to shortlist the independent variables having maximum variability amongst thousands of columns its Dimensionality Reduction problem

Dimensionality reduction i.e, reducing the number of variables that can actually provide the still gives you comparable accuracy when it comes to building prediction models.

1. randomized PCA i.e, Principal Component Analysis

2. if it doesnt work and samples > 10,000 choose a kernel approximation algorithm

3. if samples < 10,000 then Isomap or Spectral Embedding
4. if none works, pick LLE i.e, Local linear embedding algorithm.
