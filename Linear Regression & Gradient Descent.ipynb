# Linear Regression

1. We want to find variables that best fit the line/ equation of regression. Linear regression usually works on conitnuous data and estimate real values/continuous values for the dependent variable i.e the target


2. Only used for regression


3. Linear Regression is mainly of two types: Simple Linear Regression( 1 independent variable) and Multiple Linear Regression( More than one independent variable)


4. Its linear in coefficient but not necessarily linear in the independent variables. Relation between dependent and independent variable is always linear. 



### Assumptions

1. Linearity: The relationship between X and the mean of Y is linear.

eg. Plot the dependent variable against each independent variable using scatter plots. This allows you to visually inspect if there's a linear relationship between the variables.

Calculate the correlation coefficient between the dependent variable and each independent variable. A high correlation (close to 1 or -1) suggests a strong linear relationship, which is favorable for linear regression.

2. Homoscedasticity: The variance of residual is the same for any value of X.


3. Independence: Observations are independent of each other.


4. Normality: For any fixed value of X, Y is normally distributed.


5. Remove Noise. Linear regression assumes that your input and output variables are not noisy. Consider using data cleaning operations that let you better expose and clarify the signal in your data. This is most important for the output variable and you want to remove outliers in the output variable (y) if possible.


6. Remove Collinearity. Linear regression will over-fit your data when you have highly correlated input variables. Consider calculating pairwise correlations for your input data and removing the most correlated.


7. Gaussian Distributions. Linear regression will make more reliable predictions if your input and output variables have a Gaussian distribution. You may get some benefit using transforms (e.g. log or BoxCox) on you variables to make their distribution more Gaussian looking.


8. Rescale Inputs: Linear regression will often make more reliable predictions if you rescale input variables using standardization or normalization.



### How does linear regression work

We want to find the regression line that is closest to all possible points.

Equation of the line with x_0 = 1 and j = number of features:

$$ 
h_{\theta}(x^{(i)})  = (\theta_j \cdot x_j +....+ \theta_2 \cdot x_2 + \theta_1 \cdot x_1 + \theta_0 \cdot x_0) + \epsilon_i
$$

Here, the $\theta_1$ (slope) it’s are the parameters (also called weights) $\theta_0$ is the y-intercept and Єi is the random error term whose role is to add bias. The above equation is the linear equation that needs to be obtained with the minimum error. $Y_{\text{predicted}}$ is out hypothesis function $ h_{\theta}$

We need to optimize the selection of $\theta_1$ and $\theta_0$

## Cost Function and  Loss Function

Pointwise loss - Dealing with single training sample, we use the squared error loss as:

$$
LSE (\hat{y},y) = \frac{1}{2} (\hat{y}-y)^2
$$
where,
- $(\hat{y}-y)$ is the residual
- 1/2 is used for calculation convenience

Linear regression uses **Least Squared Error/ Squared error loss as loss function** ( error between predicted values and expected value). The cost function gives a convex graph and then we can complete the optimization by finding its vertex as global minimum. Its cost function is also known as **Mean Squared Error(MSE)**

For supervised learning, models are optimized by finding optimal coefficients that minimize cost function. **Cost function is the sum of losses from each data point** calculated with loss function.

$$
MSE = \frac{1}{2n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2
$$

Cost function is a parabola curve that represents the discrepancy between the predicted output of the model and the actual output.

## Gradient Descent

An optimization algorithm used in machine learning to minimize the cost function by iteratively adjusting parameters in the direction of the negative gradient, aiming to find the optimal set of parameters.

The goal of gradient descent is to find the set of parameters that minimizes this discrepancy and improves the model’s performance. The algorithm operates by calculating the gradient of the cost function and moves in the opposite direction of the gradient, known as the negative gradient direction.

We need to identify the global minima (vertex). We use **gradient descent** which takes steps based on a learning rate α, moving towards the vertex of the parabola to find the global minimum of cost function, which is also known as the point with the slope equals to 0 (not exact 0 but close to 0). 

1. Compute the gradient (slope): So to get the slope, we take the first derivative of cost function at each coefficient θ:

$$
\frac{\partial J}{\partial \theta_j} = \frac{1}{n} \sum_{i=1}^{n} (h_{\theta}(x^{(i)}) - y_i) \cdot x^{(i)}_j
$$

2. Make a step (move) in the direction opposite to the gradient, opposite direction of slope - We update $\theta$ across iterations to reduce the cost function until it reaches horixintal line.

$$
\theta_j^{(t+1)} = \theta_j^{t} - \alpha \frac{\partial J}{\partial \theta_j}
$$

where,

- $\alpha$ (Learning rate) - a hyperparameter, determines the step size taken in each iteration, influencing the speed and stability of convergence.

#### Applications:

Gradient descent can be applied to various machine learning algorithms, including linear regression, logistic regression, neural networks, and support vector machines. It provides a general framework for optimizing models by iteratively refining their parameters based on the cost function.

### Types of Gradient Descent:

The choice of gradient descent algorithm depends on the problem at hand and the size of the dataset

1.Batch Gradient Descent: Batch gradient descent updates the model’s parameters using the gradient of the entire training set. It calculates the average gradient of the cost function for all the training examples and updates the parameters in the opposite direction. 

    - Useful for Small dataset.
    - Computationally expensive and slow for large datasets.
    - Batch gradient descent guarantees convergence to the global minimum,

2.Stochastic Gradient Descent: Stochastic gradient descent updates the model’s parameters using the gradient of one training example at a time. It randomly selects a training example, computes the gradient of the cost function for that example, and updates the parameters in the opposite direction.

    - Stochastic gradient descent is computationally efficient and can converge faster than batch gradient descent.
    - Large datasets
    - Can be noisy and may not converge to the global minimum.

3.Mini-Batch Gradient Descent: Mini-batch gradient descent updates the model’s parameters using the gradient of a small subset of the training set, known as a mini-batch. It calculates the average gradient of the cost function for the mini-batch and updates the parameters in the opposite direction.

    - Most commonly used
    - Combines advantages of both
    - It is computationally efficient and less noisy than stochastic gradient descent, while still being able to converge to a good solution.

### Learning Rate

It is a constant value that decides the size of the step we take.

It must be chosen carefully to end up with local minima.

1. Learning rate is optimal, model converges to the minimum
2. Learning rate is too small, it takes more time but converges to the minimum and training is long
3. Learning rate is higher than the optimal value, it overshoots but converges ( 1/C < η <2/C)
4. Learning rate is very large, it overshoots and diverges, moves away from the minima, performance decreases on learning

#### Local Minima

The cost function may consist of many minimum points. The gradient may settle on any one of the minima, which depends on the initial point (i.e initial parameters(theta)) and the learning rate. Therefore, the optimization may converge to different points with different starting points and learning rate.

### Challenges:

1.Convergence to Local Minima if cost function has multiple peaks and valleys.

2.The choice of learning rate can significantly impact the performance of gradient descent.

3.Gradient descent can overfit the training data if the model is too complex or the learning rate is too high.

4.The convergence rate of gradient descent can be slow for large datasets or high-dimensional spaces,

5. gradient descent to get stuck in a plateau (sadddle point) in high dimension places

### To overcome these challenges:

Variations of gradient descent agorithm have been developed, such as adaptive learning rate methods, momentum-based methods, and second-order methods.
